```python
import pandas as pd
from pandasql import sqldf
import phylopandas as ph
import csv
```

```python
# reading fasta file
df_ncRna = ph.read_fasta('ncrna_NONCODE[v3.0].fasta')
df_ncRna.head()
```

```python
print(df_ncRna.shape)
```

```python
# deleting first row
df_ncRna_c=df_ncRna
df_ncRna_c=df_ncRna_c.drop(0,axis=0)
df_ncRna_c.head()
```

```python
#formating the fasta data to csv.
#creating empty dataframe
df_LncRna = pd.DataFrame(columns=['nncid','accn','class','organism','name','ref','transcriptID','url','cpcScore','cnci','sequence'])
df_LncRna['nncid']=df_ncRna_c['id']
df_LncRna['sequence']=df_ncRna_c['sequence']
df_LncRna.head()
```

```python
df_split = df_ncRna_c["description"].str.split("|", expand = True) 
df_split.head(10)
```

```python
df_LncRna["accn"]=df_split[1]
df_LncRna["class"]=df_split[2]
df_LncRna["organism"]=df_split[3]
df_LncRna["name"]=df_split[4]
df_LncRna["ref"]=df_split[5]
df_LncRna["transcriptID"]=df_split[6]
df_LncRna["url"]=df_split[7]
df_LncRna["cpcScore"]=df_split[8]
df_LncRna["cnci"]=df_split[9]
df_LncRna.head(10)
```

```python
df_LncRna.to_csv('Data/np_v2_NonCode_Seq_all.csv')
```
